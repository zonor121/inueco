```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for a simple dashboard component
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [100, 120, 150, 130, 160],
    'Expenses': [70, 80, 90, 85, 95]
}
df = pd.DataFrame(data)

# Create a simple plot
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='x', label='Expenses')
plt.title('Monthly Sales and Expenses Dashboard Component')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()

# Display some basic data as well
print("\nMonthly Data:")
print(df)
```
